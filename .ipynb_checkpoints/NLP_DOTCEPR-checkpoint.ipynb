# Coleta de dados

```python
# Step 1: Import all the important modules and packages

# for get the pdf files or url
import requests

# for tree traversal scraping in webpage
from bs4 import BeautifulSoup

# for input and output operations
import io

# For getting information about the pdfs
import PyPDF2
```

```python
# Step 2: Passing the URL and make an HTML parser with the help of BeautifulSoup.

# website to scrap
url = "https://www1.tce.pr.gov.br/conteudo/diario-eletronico-2950-2023-de-29-de-marco-de-2023/346798/area/10"

# get the url from requests get method
read = requests.get(url)

# full html content
html_content = read.content

# Parse the html content
soup = BeautifulSoup(html_content, "html.parser")
```

```python
# Step 3: We need to traverse through the PDFs from the website.

# created an empty list for putting the pdfs

list_of_links = []

for link in soup.find_all('a'):
    if 'href' in link.attrs:
        list_of_links.append(link.attrs['href'])
        
# Set is used because it never repeats the same-named elements. 
# And automatically get rid of duplicates.
set_of_pdfs = set()

for link in list_of_links:
    if link[-4:] == '.pdf':
        set_of_pdfs.add(link)
        
pdfs = list(set_of_pdfs)
```

```python
for pdf in pdfs:
    print(pdf)
```

```python
# Step 3: Create an empty python dict to save the pdf information (page and text)
information = {}

def info(pdf_path):
 
    
    # used get method to get the pdf file
    response = requests.get(pdf_path)
 
    # response.content generate binary code for
    # string function
    with io.BytesIO(response.content) as f:
 
        # initialized the pdf
        pdfReader = PyPDF2.PdfFileReader(f)
 
        # all info about pdf
        
        number_of_pages = pdfReader.numPages
        
        for page in range(number_of_pages):
            pageObj = pdfReader.getPage(page)
            information[page] = pageObj.extractText()
 
```

# Processamento de texto 

```python
import re
import nltk
from nltk.probability import FreqDist
import spacy

nlp = spacy.load("pt_core_news_sm")

def limpar_texto(texto):

  # Remover caracteres que não são letras e tokenização
    palavras =  re.findall(r'\b[A-zÀ-úü]+\b', texto.lower())

    #Remover stopwords
    stopwords = nltk.corpus.stopwords.words('portuguese')
    #Adicionando stopwords que não estão na lista do nltk 
    stopwords.append("'")
    stopwords.append("pra")
    stopwords.append("tá")
    stopwords.append('_')
    stopwords.append("art")
    stopwords.append("n")
    stopwords.append("i")
    stopwords.append("iii")
    stopwords.append("d")
    stopwords.append("s")
    stopwords.append("ii")
    stopwords.append("p")
    stopwords.append("r")
    stop = set(stopwords)

    meaningful_words = [palavra for palavra in palavras if palavra not in stopwords]
    meaningful_words_string = " ".join(meaningful_words)

    #Instanciando o objeto spacy
    spc_letras =  nlp(meaningful_words_string)

    #Lemmatização 
    tokens = [token.lemma_ if token.pos_ == 'VERB' else str(token) for token in spc_letras]

    #problemas com verbo ir
    ir = ['vou', 'vais', 'vai', 'vamos', 'ides', 'vão']
    tokens = ['ir' if token in ir else str(token) for token in tokens]
    
    return tokens
```

```python
# Reúne o texto de todas as páginas oficial em um único objeto com as strings

from functools import reduce

diario = reduce(lambda x, y: x + " " + y, information.values())
```

```python
texto_limpo = limpar_texto(diario)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
fd = nltk.FreqDist(texto_limpo)
fd.plot(30,title = "Palavras x Frequência",cumulative=False)
plt.show()
```

```python
# contagem de frequencia
fd = FreqDist(texto_limpo)
print("20 palavras mais frequentes:")
print(fd.most_common(20))
```

```python
#!pip install wordcloud

from wordcloud import WordCloud 

text = " ".join(texto_limpo)
wordcloud = WordCloud(background_color="white",width=2000, height=800, collocations = False).generate(text)
fig = plt.figure(figsize=(20,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

#### Concordância
Essa função serve para nos mostrar em quais contextos determinada palavra aparece, ela nos retorna o contexto que cerca o argumento. Isso é feito colando a palavra no centro das frases onde ela tende a aparecer através da função .concordance, que por sua vez recebe a palavra como argumento.

```python
diario_text = nltk.Text(texto_limpo)
diario_text.concordance('ressalvar')
```

#### Similaridade
Usando a função ‘similar’ temos uma lista de palavras que tendem a ocorrer no mesmo contexto. Nesse caso, o contexto são apenas as palavras que ocorrem frequentemente de qualquer lado da palavra.

```python
diario_text.similar('relator')
```

#### Bigramas
Encontrar bigramas é uma parte importante para a análise de um texto. Mas primeiro, o que são bigramas? São palavras no texto que possuem um sentido único estando juntas, podendo até ocorrer separadas, mas com um sentido quando juntas. Por exemplo, nomes compostos (João Alfredo, Maria Júlia) ou nomes de cidades (São Paulo, Rio de Janeiro, Nova Iorque) são exemplos claros de bigramas, essas palavras podem ocorrer sozinhas, mas ocorrendo juntas possuem um significado.

Para encontrar os bigramas de um texto com NLTK, basta chamar a função ‘collocations’:

```python
diario_text.collocations() 
```

#### Gráfico de dispersão
Essa função serve para analisar a ocorrência das palavras ao longo de todo o texto — no nosso caso, do livro — e mostra nos riscos do eixo horizontal quando essa palavra tende a ocorrer nele

```python
plt.figure(figsize=(15, 10))
diario_text.dispersion_plot(['honorários', 'despesas', 'repasses',
                            'convênio', 'itambaracá'])
```
